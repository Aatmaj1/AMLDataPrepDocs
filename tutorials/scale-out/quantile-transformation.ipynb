This notebook is a continuation from the local quantile transformation reference notebook. In this notebook, we will demonstrate the operationalization of the previously saved package using our DrepEngine and on Spark.

Let's first download our previously saved package.

```python
from tempfile import mkdtemp
from os import path
from urllib.request import urlretrieve

package_root = mkdtemp()
package_path = path.join(package_root, "quantile_transform.dprep")
print("Downloading package to: {}".format(package_path))
urlretrieve("https://dpreptestfiles.blob.core.windows.net/testfiles/quantile_transform.dprep", package_path)
```

Let's get the Dataflow from the package that contains our previous quantile transformations.

```python
import azureml.dataprep as dprep

package = dprep.Package.open(package_path)
df = package.dataflows[0]
```

Let's grab the profile using the DPrepEngine. The results looks identical from our previous run when we created this dataflow.

```python
df.get_profile()
```

Let's now get the data profile again, but this time, let's convert our Dataflow to a spark DataFrame and use DataFrame's `describe` method to get the statistics. As you can see, the results is almost identical to our local execution.

```python
spark_df = df.to_spark_dataframe().describe().show()
```
