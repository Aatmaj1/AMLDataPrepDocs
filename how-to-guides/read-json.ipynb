```python
import azureml.dataprep as dprep
```

In addition to the ability to load text files DataPrep also can load JSON files.

```python
dataflow = dprep.read_json(path='./data/json.json')
dataflow.head(10)
```

When you use read_json, DataPrep will attempt to extract data from the file into a table. 
You can also control file encoding DataPrep should use as well as whether DataPrep should flatten nested JSON arrays.
Choosing the option to flatten nested arrays could produce much bigger number of rows.

```python
dataflow = dprep.read_json(path='./data/json.json', flatten_nested_arrays=True)
dataflow.head(10)
```

```python
df = dataflow.to_pandas_dataframe()
df
```
